# Intro:
**Overfitting is an important deal to machine learning. L1(Lasso) and L2(Ridge) regularizations are used to overcome this issue.**


 


```python
#Libraries:
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sbn
```

```python
#Supress warnings for clean notebook
import warnings
warnings.filterwarnings("ignore")
```

```python
#Read the dateset
dataset = pd.read_excel("Melbourne_housing_FULL.xlsx")
dataset.head()
```

```python
dataset.nunique()
```

```python
# let's use limited columns which makes more sense for serving our purpose
columns_to_use = ["Suburb", "Rooms", "Type", "Method", "SellerG", "Regionname", "Propertycount",
                 "Distance", "CouncilArea", "Bedroom2", "Bathroom", "Car", "Landsize", "BuildingArea", "Price"]
dataset = dataset[columns_to_use]
dataset.head()
```

```python
dataset.shape
```

# Checking for NaN Values:

```python
dataset.isna().sum()
```

**Handling Missing Values:**

```python
# Some feature's missing values can be treated as zero (another class for NA values or absence of that feature)
# like 0 for Propertycount, Bedroom2 will refer to other class of NA values
# like 0 for Car feature will mean that there's no car parking feature with house
columns_to_fill_zero = ['Propertycount', 'Distance', 'Bedroom2', 'Bathroom', 'Car']
dataset[columns_to_fill_zero] = dataset[columns_to_fill_zero].fillna(0)

# other continuous features can be imputed with mean for faster results since our focus is on Reducing overfitting
# using Lasso and Ridge Regression
dataset['Landsize'] = dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
```

**Drop NA values of price, since it is our predictive variable we will not impute it.**

```python
dataset.dropna(inplace=True)
```

```python
dataset.shape
```

**One hot encoding for the categorical features**

```python
dataset = pd.get_dummies(dataset, drop_first=True)
dataset.head()
```

**Bifurcate the dataset into train and test:**

```python
X = dataset.drop('Price', axis=1)
y = dataset["Price"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(X_train, y_train)
```

```python
reg.score(X_test, y_test)
```

```python
reg.score(X_train, y_train)
```

***Here training score is 68% but test score is 13.8% which is very low.***
***Normal Regression clearly averfitting the data. Hence, other models should be tried.***

# Lasso (L1 Regularized) Regression Model:

```python
from sklearn import linear_model
lasso_reg = linear_model.Lasso(alpha=50, max_iter=100, tol=0.1)
lasso_reg.fit(X_train, y_train)
```

```python
lasso_reg.score(X_test, y_test)
```

```python
lasso_reg.score(X_train, y_train)
```

# Ridge (L2 Regularized) Regression Model:

```python
from sklearn.linear_model import Ridge
ridge_reg= Ridge(alpha=50, max_iter=100, tol=0.1)
ridge_reg.fit(X_train, y_train)
```

```python
ridge_reg.score(X_test, y_test)
```

```python
ridge_reg.score(X_train, y_train)
```

**We see that Lasso and Ridge Regularizations prove to be beneficial when our Simple Linear Regression Model overfits. These results may not be that contrast but significant in most cases.Also that L1 & L2 Regularizations are used in Neural Networks too**
