```python
import pandas as pd
import numpy as np
df = pd.read_excel("homeprices.xlsx")
df
```

```python
pd.get_dummies(df.town)
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df,dummies], axis='columns')
merged
```

```python
#we can't use the town column for linear regression.
#First drop the town column.
#Second drop one of the dummy variables column. Because the last category will automatically selected as positive.
#axis=1 means that we are gonna drop from the columns. 0=horizontal, 1=vertical
final = merged.drop(['town', 'west windsor'], axis=1)
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
#y = price, except that all of them are features of x
x = final.drop('price', axis=1)
x
```

```python
y = final.price
y
```

```python
#Train the program:
model.fit(x, y)
```

```python
#area:2800, town:west windsor
model.predict([[2800, 0,0]])
```

```python
#area:5000, town:monroe township
model.predict([[5000, 1,0]])
```

```python
#How accurate the model: 95% is accurate
model.score(x,y)
```
