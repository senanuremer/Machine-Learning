**How to determine proper number of clusters?**
 

**When you increase the cluster the error will decrease:**


**GOOD KEY VALUE IS 4 BY ELBOW TECHNIQUE**


```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel("income.xlsx")
df.head()
```

```python
plt.scatter(df["Age"], df["Income($)"])
```

```python
km = KMeans(n_clusters=3)
km
```

```python
y_predicted = km.fit_predict(df[["Age", "Income($)"]])
y_predicted
```

```python
df["cluster"] = y_predicted
df.head()
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt.scatter(df1.Age, df1["Income($)"], color="green")
plt.scatter(df2.Age, df2["Income($)"], color="red")
plt.scatter(df3.Age, df3["Income($)"], color="blue")

plt.xlabel("Age")
plt.ylabel("Income($)")
plt.legend()
```

```python
scaler = MinMaxScaler()
scaler.fit(df[["Income($)"]])
df["Income($)"] = scaler.transform(df[["Income($)"]])
df
```

```python
scaler = MinMaxScaler()
scaler.fit(df[["Age"]])
df["Age"] = scaler.transform(df[["Age"]])
df
```

```python
km = KMeans(n_clusters=3)
y_predicted = km.fit_predict(df[["Age", "Income($)"]])
y_predicted
```

```python
df["cluster"] = y_predicted
df
```

```python
#Centroids of the clusters:
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]

plt.scatter(df1.Age, df1["Income($)"], color="green")
plt.scatter(df2.Age, df2["Income($)"], color="red")
plt.scatter(df3.Age, df3["Income($)"], color="blue")
plt.legend()
```

```python
#Elbow plot:
k_rng = range(1,10)
sse = []
for k in k_rng:
    km = KMeans(n_clusters=k)
    km.fit(df[["Age", "Income($)"]])
    #sum of error:
    sse.append(km.inertia_)
```

```python
sse
```

```python
plt.xlabel("K")
plt.ylabel("Sum of squared error")
plt.plot(k_rng,sse)
```

From the graph above, 3 is the proper number of clusters!
