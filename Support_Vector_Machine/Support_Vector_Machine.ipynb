```python
import pandas as pd
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
#Look what is the basic features of iris:
dir(iris)
```

```python
iris.feature_names
```

```python
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df.head()
```

```python
df["target"] = iris.target
df.head()
```

```python
iris.target_names
```

From the above we can observe that zeroth index is setosa, first index is versicolor, second index is virginia. In the table above, we see that the target index is zero, that is, the data is data related to Setosa.

```python
#Let's visualize target_name: versicolor (index=1)
df[df.target==1].head()
```

Data regarding Versicolor starts from line 50.

```python
##Let's visualize target_name: virginica (index=2)
df[df.target==2].head()
```

Data regarding Virginica starts from line 100.

```python
#Add a new column to df that called as flower_name.
#Take the flower name from the target_name feature of iris:
df["flower_name"] = df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
df1.head()
```

```python
#Visualizing the df0-df1 to make comparison with the sepal length and sepal width:
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.scatter(df0['sepal length (cm)'], df0["sepal width (cm)"], color ="red", marker="+")
plt.scatter(df1['sepal length (cm)'], df1["sepal width (cm)"], color ="blue", marker=".")
```

```python
#Visualizing the df0-df1 to make comparison with the petal length and petal width:
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")
plt.scatter(df0['petal length (cm)'], df0["petal width (cm)"], color ="red", marker="+")
plt.scatter(df1['petal length (cm)'], df1["petal width (cm)"], color ="blue", marker=".")
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.drop(["target", "flower_name"], axis=1)
X.head()
```

```python
y = df.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
#Support vector machine (svm:model, svc:classifier)
from sklearn.svm import SVC
model = SVC()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
